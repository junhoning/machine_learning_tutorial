```python
df_sex = df.Sex
```

```python
sx_dummies = pd.get_dummies(df_sex, prefix='sex')
sx_dummies.head()
```

```python
sx_dropped = df.drop('Sex', axis=1)

new_df = sx_dropped.join(sx_dummies)
new_df.head()
```

```python
embark = df.Embarked
embark.head()
```

```python
embark_dummies * 3
```

```python
embark_dummies = pd.get_dummies(embark, prefix='embark')

new_df = new_df.drop('Embarked', axis=1)

new_df = new_df.join(embark_dummies)
new_df.head()
```

```python
new_df = new_df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
new_df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Dummies ì²˜ë¦¬ ì•ˆ í•˜ê³  String ìžˆëŠ” ê±´ ë‹¤ ëº€ ìžë£Œ
org_df = df.drop(['Name', 'Sex', 'Ticket', 'Fare', 'Cabin', 'Embarked'], 1)
org_df = org_df.dropna()
num_dataset = int(0.8 * len(org_df))

trainset = org_df[:num_dataset]
testset = org_df[num_dataset:]

model = RandomForestClassifier()
model.fit(trainset[trainset.columns[1:]], trainset.Survived)

preds = model.predict(testset[trainset.columns[1:]])
np.mean(preds == testset.Survived)
```
