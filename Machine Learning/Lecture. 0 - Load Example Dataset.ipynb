```
from sklearn import datasets
import pandas as pd
import numpy as np
```

```
def load_to_pd(dataset):
    df = pd.DataFrame(data = np.c_[dataset['data'], dataset['target']],
                  columns = dataset['feature_names'] + ['target'])
    return df
```

```
df = load_to_pd(datasets.load_iris())
df.head()
```

```
df = load_to_pd(datasets.load_diabetes())
df.head()
```

```
df = load_to_pd(datasets.load_wine())
df.head()
```
