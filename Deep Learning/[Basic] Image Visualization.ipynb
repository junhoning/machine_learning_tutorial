## package 불러오기

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

---

## 이미지 파일 준비

opencv로도 열수 있지만, shape 순서가 바뀔 때도 있어서 PIL로 여는 것을 개인적으로 선호

```python
path = 'source/tiger.jpg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

---

## 이미지 열기

```python
image.shape
```

```python
np.max(image), np.min(image)
```

이미지를 열기 전에 shape 및 min, max를 통해서 이미지의 range 확인 필요 

___

## 그래프로 시각화 하기

```python
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

---

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

---
## 이미지 흑백으로 열기

```python
image_bw = image_pil.convert('L')
image_bw = np.array(image_bw)
```

---
## 흑백 이미지 열기

```python
plt.imshow(image_bw)
plt.show()
```

---
### 다른 색상으로 cmap 표현하기

### gray scale

```python
plt.imshow(image_bw, cmap='gray')
plt.show()
```

---

### RdBu

```python
plt.imshow(image_bw, 'RdBu')
plt.show()
```

---

### jet

```python
plt.imshow(image_bw, 'jet')
plt.show()
```

---
## Colorbar 추가하기

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

# 두번째 이미지 열기

```python
path = 'source/kitten.jpg'

image_pil = Image.open(path)
cat_image = np.array(image_pil)
```

---
## 두번쨰 이미지를 첫번째 이미지 모양에 맞추기

```python
cat_image.shape
```

```python
plt.imshow(cat_image)
plt.show()
```

---

## 두번째 이미지 합치기 위한 준비

```python
import cv2

cat_image = cv2.resize(cat_image, \
                       (image.shape[1], image.shape[0]))

image.shape
```

```python
cat_image.shape
```

---

## Image 합치기

```python
plt.imshow(image, alpha=0.8)
plt.imshow(cat_image, alpha=0.5)
plt.show()
```
